## Ames Iowa Dataset

```python
# import the python libraries
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
# read in a dataset
df = pd.read_csv('./data/iowa.csv')
```

```python
# take a look at the data
df.head()
```

```python
# scatter plot
fig = px.scatter(df, x="1stFlrSF",
                 y="SalePrice", 
                 trendline="ols",
                 trendline_color_override="red"
                )
fig.show()
```

```python
# scatter plot
fig = px.scatter(df, 
                 y="SalePrice", 
                 x="2ndFlrSF",
                 trendline="ols",
                 trendline_color_override="red"
                )
fig.show()
```

```python
# save your image output
fig.write_html('price_prediction.html')
```
